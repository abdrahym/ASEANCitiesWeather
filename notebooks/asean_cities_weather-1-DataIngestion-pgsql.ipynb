# ASEAN Cities Weather Data

[ASEAN Cities Weather Data 2010-YTD(https://www.kaggle.com/datasets/bwandowando/asean-cities-weather-data-2010-2024)](https://www.kaggle.com/datasets/bwandowando/asean-cities-weather-data-2010-2024)

## About Dataset



### Context

Coordinates of ASEAN cities from [https://simplemaps.com/](https://simplemaps.com/), hourly and daily Weather Dataset from **[https://open-meteo.com/](https://open-meteo.com/)** from January 01, 2010 to November 05, 2024.

### Field documentation go here

-   **[https://open-meteo.com/en/docs/historical-weather-api#start\_date=2010-01-01&end\_date=2025-04-05](https://open-meteo.com/en/docs/historical-weather-api#start_date=2010-01-01&end_date=2025-04-05)**
-   Datetime is in GMT

### Visually verify coordinates

-   Replace the placeholders with the latitude and longitude values of the cities **[https://www.latlong.net/c/?lat=(latitude)&long=(longitude](https://www.latlong.net/c/?lat=(latitude)&long=(longitude))**
-   Another way to visualize the coordinates **[https://www.ncei.noaa.gov/maps/radar/](https://www.ncei.noaa.gov/maps/radar/)**
